I went to the Smithhaven Trader Joe's location and collected raw data for this project

I collected data about: 
Number of customers entering within a 5 minute period 
- 15 samples during on-peak hours
- 15 samples during off-peak hours

Number of item removals off the shelf within a 5 minute period 
- Item categories are: frozen, snacks, and bakery
- 15 samples during on-peak hours
- 15 samples during off-peak hours

# Data Exploration #

```python
# All necessary imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# Data collected for on-peak hours
on_peak_data = {
    'Customers Entering': [50, 39, 55, 40, 35, 46, 34, 42, 50, 31, 50, 44, 38, 36, 52],
    'Frozen Items': [27, 17, 56, 48, 26, 31, 42, 12, 49, 47, 61, 61, 14, 27, 53],
    'Snack Items': [34, 29, 27, 32, 39, 38, 31, 28, 37, 23, 33, 30, 33, 36, 29],
    'Bakery Items': [10, 6, 6, 5, 7, 6, 11, 8, 6, 8, 6, 9, 9, 4, 8]
}

# Convert the on-peak data into a pandas DataFrame
on_peak_df = pd.DataFrame(on_peak_data)

# Descriptive statistics for on-peak data
on_peak_descriptive_stats = on_peak_df.describe()

# Round to 3 decimal points
on_peak_descriptive_stats_rounded = on_peak_descriptive_stats.round(3)

# Print the rounded descriptive statistics as a single table
print(on_peak_descriptive_stats_rounded.to_string())
```

```python
# Data collected for off-peak hours 
off_peak_data = {
    'Customers Entering': [15, 29, 21, 18, 22, 28, 17, 21, 16, 22, 19, 16, 25, 17, 25],
    'Frozen Items': [18, 15, 14, 16, 15, 13, 12, 16, 14, 14, 16, 15, 15, 15, 13],
    'Snack Items': [20, 32, 8, 25, 33, 3, 3, 24, 35, 37, 29, 13, 22, 37, 19],
    'Bakery Items': [5, 3, 1, 0, 2, 2, 6, 2, 4, 0, 5, 0, 5, 4, 1]
}

# Convert the off-peak data into a pandas DataFrame
off_peak_df = pd.DataFrame(off_peak_data)

# Descriptive statistics for off-peak data
off_peak_descriptive_stats = off_peak_df.describe()

# Round to 3 decimal points
off_peak_descriptive_stats_rounded = off_peak_descriptive_stats.round(3)

# Print the rounded descriptive statistics as a single table
print(off_peak_descriptive_stats_rounded.to_string())
```

# Plot Histograms #

```python
# Plot histograms for each category of data
categories = ['Customers Entering', 'Frozen Items', 'Snack Items', 'Bakery Items']

fig, axes = plt.subplots(4, 2, figsize=(10, 15))  # 4 rows, 2 columns
fig.tight_layout(pad=5.0)

for i, category in enumerate(categories):
    # Plot on-peak data histogram
    axes[i, 0].hist(on_peak_df[category], bins=10, color='blue', edgecolor='black', alpha=0.7)
    axes[i, 0].set_title(f'On-Peak: {category}')
    axes[i, 0].set_xlabel(category)
    axes[i, 0].set_ylabel('Frequency')

    # Plot off-peak data histogram
    axes[i, 1].hist(off_peak_df[category], bins=10, color='green', edgecolor='black', alpha=0.7)
    axes[i, 1].set_title(f'Off-Peak: {category}')
    axes[i, 1].set_xlabel(category)
    axes[i, 1].set_ylabel('Frequency')

plt.show()
```
